```python
from collections import Counter
import nltk
import string
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
text=input()
text=text.lower()
for t in string.punctuation:
    text=text.replace(t,"")
stop_words=set(stopwords.words('english'))
words=word_tokenize(text)
print("Tokenized words:",words)
total_words=len(words)
unique_words=set(words)
freq=Counter(words)
print("\nWord Frequency:",freq)
print("Vocabulary Size:",len(unique_words))
print("\n Top 5 most Frequent words:")
for word,count in freq.most_common(5):
    print(word,":",count)
ttr=len(unique_words)/total_words
print("\nType token Ratio:",round(ttr,2))
filtered_words=[w for w in words if w not in stop_words]
filtered_freq=Counter(filtered_words)
print("Words after removing Stop Words:",filtered_words)
print(freq.most_common(1)[0])
print("Most Dominant topic:",word)
print("Frequency:",count)
word_list=list(filtered_freq.keys())
counts_list=list(filtered_freq.values())
plt.figure(figsize=(10,8))
plt.bar(word_list,counts_list)
plt.title("News Article Analysis")
plt.xlabel('Word list')
plt.ylabel('Frequency')
plt.xticks(rotation=100)
plt.show()
```
