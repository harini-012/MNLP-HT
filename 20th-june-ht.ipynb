```python
'''
Date: 20-06-2026
Home Task
Scenario: Smart News Analysis System 
A media company receives thousands of news articles every day. To help journalists quickly 
understand the content, the company wants to build a Smart News Analysis System using NLTK. 
The system should analyze the news article and perform various NLP tasks using different NLTK 
modules. 
Tasks 
1. Parse the sentence structure to identify grammatical relationships using nltk.parse.  
2. Extract named entities such as persons, organizations, and locations using nltk.chunk.  
3. Classify the news article into categories such as Sports, Politics, Business, or Technology using 
nltk.classify.  
4. Evaluate the classification model using accuracy and precision measures with nltk.metrics.  
5. Calculate word frequencies and probability distributions of important terms using 
nltk.probability.  
6. Analyze semantic meaning and relationships between important words using nltk.sem.  
7. Generate bigrams and trigrams from the article using nltk.util.
'''
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk import pos_tag
 
from nltk.chunk import ne_chunk,RegexpParser
from nltk.classify import NaiveBayesClassifier
from nltk.metrics import precision
from nltk.probability import FreqDist
from nltk.util import bigrams,trigrams
from nltk.corpus import wordnet as wn
from collections import defaultdict
from nltk.classify.util import accuracy
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('words')
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')

nltk.download('wordnet')
news="""
Microsoft announced a new AI platform in Chennai.
The company expects the technology to improve business productivity.
"""
tokens=word_tokenize(news)
pos_tags=pos_tag(tokens)
print("PARSING:")
grammar="""NP: {<DT>?<JJ>*<NN.*>+}
VP: {<VB.*><NP|PP>*}
"""
parser=RegexpParser(grammar)
parse_tree=parser.parse(pos_tags)
print(parse_tree)
print("Named Entites:")
ner_tree=ne_chunk(pos_tags)
for chunk in ner_tree:
    if hasattr(chunk,"label"):
        entity=" ".join(word for word,tag in chunk.leaves())
        if chunk.label()=="GPE":
            print("Location:",entity)

        elif chunk.label() in ["ORGANIZATION", "PERSON"] and entity == "Microsoft":
            print("Organization:", entity)

        else:
            print(chunk.label(),":",entity)
        
print("Classification:")
train_data=[({"word": "cricket"}, "Sports"),
    ({"word": "football"}, "Sports"),
    ({"word": "election"}, "Politics"),
    ({"word": "minister"}, "Politics"),
    ({"word": "market"}, "Business"),
    ({"word": "profit"}, "Business"),
    ({"word": "technology"}, "Technology"),
    ({"word": "ai"}, "Technology")
           ]
classifier=NaiveBayesClassifier.train(train_data)
category_votes=[]
for word in tokens:
    category_votes.append(classifier.classify({"word":word.lower()})
                         )
    predicted_category=max(set(category_votes),key=category_votes.count)
print("Category:",predicted_category)
print("Model Evaluation")
test_data = [
    ({"word": "ai"}, "Technology"),
    ({"word": "technology"}, "Technology"),
    ({"word": "profit"}, "Business"),
    ({"word": "cricket"}, "Sports")
]
acc=accuracy(classifier,test_data)
reference=defaultdict(set)
predicted=defaultdict(set)
for i,(features,label) in enumerate(test_data):
        reference[label].add(i)
        predicted[classifier.classify(features)].add(i)
        tech_precision=precision(reference["Technology"],predicted["Technology"])
print("Accuracy:",round(acc*100,2),"%")
print("Technology Precision:",round(tech_precision*100,2),"%")
print("Word Frequency:")
words=[w.lower() for w in tokens if w.isalpha()]
fdist=FreqDist(words)
for word,freq in fdist.items():
        print(word,":",freq)
        print("Probablity Distribution")
        N=len(words)
for word in fdist:
        print(f"P({word})={fdist[word]}/{N}={fdist[word]/N:.2f}")
        print("Semantic Analysis")
for word in ["technology","business"]:
    synsets=wn.synsets(word)
    if synsets:
        print(word,"->",synsets[0].definition())
s1=wn.synsets("technology")[0]
s2=wn.synsets("business")[0]
print("Similarity:",round(s1.wup_similarity(s2),2))
print("Bigrams")
for bg in bigrams(words):
    print(bg)
    print("Trigrams")
for tg in trigrams(words):
    print(tg)
        
```
